### Datasets ###

1. A small version of the [GTZAN](http://marsyasweb.appspot.com/download/data_sets/) music genre dataset. The goal is to classify different genres of music. The genre labels are:

    0 'blues'
    
    1 'classical'
    
    2 'country'
    
    3 'disco'
    
    4 'hiphop'
    
    5 'jazz'
    
    6 'metal'
    
    7 'pop'
    
    8 'reggae'
    
    9 'rock'
    
    Get the small dataset version that we will be using here: [google drive link](https://drive.google.com/open?id=0B1gxPy-6qVriblhlYTNFS2tXd2s)
    
    A larger version of the dataset can be found here: [google drive link](https://drive.google.com/open?id=0B0NEMdSW1_-IZnM5cVRSWnk5cVk). There are 8000 labeled training examples, 100 validation examples, and 100 test examples. We don't recommend implementing cross-validation with this dataset, as the training data may contain highly correlated datapoints.<br/><br/>
   
    
2. We will use this lyrics data to carry out semantics analysis. Download [songlyrics.csv](https://www.kaggle.com/mousehead/songlyrics) from this link.
        
Larger datasets are available upon request for you to use for your final project.
