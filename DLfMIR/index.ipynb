### Datasets ###

We will be working with four datasets, the first one consists of fake data that we will generate on-the-fly in Julia and Python. The other three are available in these links, and you will use them in your Cloud9 workspace.

WARNING: The free Cloud9 memory space is very small (2GB). Therefore, we will be uploading and erasing datasets from Cloud9 as neded. Don't upload any data to Cloud9 until indicated to do so, or you may run out of memory space.

1. The first dataset we will explore is a small and pre-processed version of the [Iowa Musical Instrument Samples Dataset](http://theremin.music.uiowa.edu/index.html). The goal is to clasify the sounds of instruments based on the timbre of single pitches. The instrument labels are:

    0 'Bass'
    
    1 'Bassoon'
    
    2 'Cello'
    
    3 'Clarinet' 
    
    4 'Flute' 
    
    5 'Guitar'
    
    6 'Horn'
    
    7 'Sax'
    
    8 'Trombone'
    
    9 'Trumpet'
    
    10 'Viola'
    
    11 'Violin'
    
    Get the small, pre-processed dataset version that we will be using here: [google drive link](https://drive.google.com/open?id=0B1gxPy-6qVriV1FUWUtiZVFUV2s)


2. The second dataset we will use is a small version of the [GTZAN](http://marsyasweb.appspot.com/download/data_sets/) music genre dataset. The goal is to classify different genres of music. The genre labels are:

    0 'blues'
    
    1 'classical'
    
    2 'country'
    
    3 'disco'
    
    4 'hiphop'
    
    5 'jazz'
    
    6 'metal'
    
    7 'pop'
    
    8 'reggae'
    
    9 'rock'
    
    Get the small dataset version that we will be using here: [google drive link](https://drive.google.com/open?id=0B1gxPy-6qVriblhlYTNFS2tXd2s)
    

3. The third datset is one with lyrics. We will use this data to carry out semantics analysis. Download [songlyrics.csv](https://www.kaggle.com/mousehead/songlyrics) from this link.
        
Larger datasets are available upon request for you to use for your final project. Also, because of Cloud9's limited memory capacity, you will have to find better computational power (Most laptops, although relatively slow, do a decent job at traning neural networks).
